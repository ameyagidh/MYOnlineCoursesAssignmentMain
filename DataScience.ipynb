```python
sum_ = 0
c = 1
for d in mpg:
    if d["cty"]:
        sum_ += float(d["cty"])
        c += 1
print(sum_/c)
```

```python
sum_ = 0
c = 0
for d in mpg:
    if d["hwy"]:
        sum_ += float(d["hwy"])
        c += 1
print(sum_/c)
```

```python
cyl_u = set(d["cyl"] for d in mpg)
print(cyl_u)
```

```python
vehicleclass = set(d['class'] for d in mpg)
vehicleclass
lst = []
for veh in vehicleclass:
    sum_ = 0
    c = 0
    for each_ in mpg:
        if veh == each_["class"]:
            sum_ += float(each_["hwy"])
            c += 1
    lst.append((veh,sum_/c))
print(lst)
```

```python
with open("ferpa.txt") as file:
    wiki = file.read()
# re.findall(,grades)
wiki
```

```python
# re.findall("[a-zA-Z]{1,100}.*\[edit\]",wiki)
re.findall("[\w]{1,100}.*\[edit\]",wiki)
```

```python
re.findall("[\w].*\[edit\]",wiki)
```

```python
for i in re.findall("[\w].*\[edit\]",wiki):
    print(re.split("[\[]",i)[0])
```

```python
for i in re.finditer("([\w].*)(\[edit\])",wiki):
    print(i.group(1))
```

```python
for i in re.finditer("(?P<title>[\w].*)(?P<edit_tag>\[edit\])",wiki):
    print(i.groupdict()["title"])
```

```python
for i in re.finditer("(?P<title>[\w].*)(?=\[edit\])",wiki):
    print(i)
```
